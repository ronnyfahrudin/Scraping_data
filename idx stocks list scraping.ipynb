```python
# import selenium
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome("D:/Karir/Evidensi/chromedriver.exe")

# get the driver URL
driver.get("https://www.idx.co.id/data-pasar/data-saham/daftar-saham/")

# set table entries
## to display 100 entries
set_table_entries = driver.find_element_by_name("stockTable_length")
set_table_entries.click()
entry_options = set_table_entries.find_elements_by_tag_name("option")
entry_options[3].click()

time.sleep(5)

tables = []

# fill tables with table data
while True:
    table_data = driver.find_element_by_tag_name("tbody")

    print(table_data)

    for i in table_data.find_elements_by_tag_name("tr"):
        print(i)
        data = []

        for j in i.find_elements_by_tag_name("td"):
            data.append(j.text)

        tables.append(data)
    
    if len(table_data.find_elements_by_tag_name("tr")) < 100:
        break

    next_button = driver.find_element_by_class_name("next")
    next_button.click()
    time.sleep(4)

    
```

```python
df = pd.DataFrame(tables)
```
